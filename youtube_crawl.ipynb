```python
import os
import argparse
from pytube import YouTube
import librosa
import soundfile as sf
import re
from pytube import Playlist
```

```python
save_dir = '/content/drive/MyDrive/Data'
```

```python
number = 0
video = ['https://www.youtube.com/watch?v=dk2BgFuY6jc&t=3887s', 'https://www.youtube.com/watch?v=qygZpQdQXLM', 'https://www.youtube.com/watch?v=cmvrnvdVAdI']
for i in range(0,len(video)):
    number = number + 1
    if number == 21:
        break
    yt = YouTube(video[i])
    yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    yt.download(save_dir)
    print(number)
```

```python
for i in os.listdir(save_dir):
    os.rename(save_dir + '/' + i, save_dir + '/' + i.replace("'", "").replace('-', '_').replace(' ', '_'))
```

```python
import torch
import torchaudio
import torchaudio.functional as F
from moviepy.editor import *
import glob
```

```python
k = 56
for i in glob.glob(save_dir + '/*.mp4'):
    videoclip = VideoFileClip(i)
    audioclip = videoclip.audio
    audioclip.write_audiofile(save_dir + str(k) +'.mp3')
    k = k+1
```

```python
from pydub import AudioSegment

k = 56
for i in glob.glob('/content/drive/MyDrive/Data/*.mp3'):
    sound = AudioSegment.from_mp3(i)
    sound.export('/content/drive/MyDrive/Data/' +  str(k) + '.wav', format="wav")
    k = k + 1
```
